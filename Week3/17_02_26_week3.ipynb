```python
# Performing data manipulations, data operations, ranking on a dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url="https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df=pd.read_csv(url)

print("Original Dataset:")
print(df.head())

#FILTERING

print("\nRecords where Total_Bill>20 and Tips>3")
filtered_and=df[(df['total_bill']>20)&(df['tip']>3)]
print(filtered_and)

print("\nRecords where Day is Saturday or Sunday")
filtered_or=df[(df['day']=='Sat')|(df['day']=='Sun')]
print(filtered_or)

print("\nRecords where Smoker is NOT Yes")
filtered_not=df[~(df['smoker']=='Yes')]
print(filtered_not)

print("\nFemale and Non-Smoker and Total_Bill>15")
combined_filter=df[(df['sex']=='Female')&(df['smoker']=='No')&(df['total_bill']>15)]
print(combined_filter)

print("\nNumber of records satisfying combined condition:",len(combined_filter))

# SORTING

print("\nSorted by Total_Bill in ascending order:")
sorted_asc=df.sort_values(by='total_bill')
print(sorted_asc.head())

print("\nSorted by Tip in descending order:")
sorted_desc=df.sort_values(by='tip',ascending=False)
print(sorted_desc.head())

print("\nSorted by Day and Total_Bill")
sorted_multi=df.sort_values(by=['day','total_bill'])
print(sorted_multi.head())

#  GROUPING

print("\nAverage Total Bill by Day")
group_day=df.groupby('day')['total_bill'].mean()
print(group_day)

print("\nAverage Tip by Sex and Smoker")
group_multi=df.groupby(['sex','smoker'])['tip'].mean()
print(group_multi)

print("\nCount of Records by Day")
group_count=df.groupby('day').size()
print(group_count)

# VISUALIZATIONS

# Line Plot – Average Total Bill by Day
plt.figure()
group_day.plot(kind='line', marker='o')
plt.title("Average Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Average Total Bill")
plt.show()


# Bar Plot – Count of Records by Day
plt.figure()
group_count.plot(kind='bar')
plt.title("Count of Records by Day")
plt.xlabel("Day")
plt.ylabel("Count")
plt.show()


#Scatter Plot – Total Bill vs Tip
plt.figure()
plt.scatter(df['total_bill'], df['tip'])
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()


#Histogram – Distribution of Total Bill
plt.figure()
plt.hist(df['total_bill'], bins=20)
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()


#Box Plot – Total Bill by Day
plt.figure()
sns.boxplot(x='day', y='total_bill', data=df)
plt.title("Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()

#Pie Chart – Smoker Distribution
plt.figure()
smoker_counts=df['smoker'].value_counts()
plt.pie(smoker_counts, labels=smoker_counts.index, autopct='%1.1f%%')
plt.title("Smoker vs Non-Smoker Distribution")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create data
data = [10, 12, 15, 18, 20, 22, 25, 30, 35, 40]

# Create DataFrame
df = pd.DataFrame(data, columns=["Values"])

print("Dataset:")
print(df)

print("\nSummary Statistics:")
print(df.describe())

print("\nVariance:", np.var(data))
print("Standard Deviation:", np.std(data))

# Plot histogram
plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Data Distribution")
plt.show()

# Calculate mean and median
mean_value = np.mean(data)
median_value = np.median(data)

print("\nInference")
print("Mean Value:", mean_value)
print("Median Value:", median_value)
if mean_value > median_value:
    print("Data is slightly right-skewed")

elif mean_value < median_value:
    print("Data is slightly left-skewed")

else:
    print("Data is symmetric")
```

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load Iris dataset (working public CSV link)
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

print("Dataset (first 10 rows):")
print(df.head(10))

# Select the sepal length column
data = df["sepal_length"]

print("\nSummary Statistics:")
print(data.describe())

print("\nVariance:", np.var(data))
print("Standard Deviation:", np.std(data))

# Plot histogram
plt.figure(figsize=(8,5))
plt.hist(data, bins=10, color='skyblue', edgecolor='black')

# Mean line
plt.axvline(np.mean(data), color='red', linestyle='dashed',
            linewidth=2, label=f"Mean: {np.mean(data):.2f}")

# Median line
plt.axvline(np.median(data), color='green', linestyle='dashed',
            linewidth=2, label=f"Median: {np.median(data):.2f}")

plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Length")
plt.legend()
plt.show()

# Skewness inference
mean_value = np.mean(data)
median_value = np.median(data)

print("\nInference:")
print("Mean value:", mean_value)
print("Median value:", median_value)

if mean_value > median_value:
    print("Data is slightly right-skewed")
elif mean_value < median_value:
    print("Data is slightly left-skewed")
else:
    print("Data is symmetric")
```
