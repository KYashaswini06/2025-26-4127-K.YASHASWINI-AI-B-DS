```python
#Descriptive Statistics
import pandas as pd
data={
    "Ages":[22,25,47,35,46,50,28,33,39,41],
    "Salaries":[3000,3200,5500,4000,5200,6000,3500,3700,4500,4800]
}
df=pd.DataFrame(data)
print("Original DataFrame:")
print(df)
print("\nSummary Statistics:\n")
print(df.describe())
print("\nMean:\n",df.mean())
print("\nMedian:",df.median())
print("\nMode:\n",df.mode())
print("\nMinimum value:\n",df.min())
print("\nMaximum value:\n",df.max())
print("\nStandard Deviation:\n",df.std())
print("\nCount:\n",df.count())
print("\nVariance:\n",df.var())
print("\nSkewness:\n",df.skew())
print("\nKurtosis:\n",df.kurt())
```
